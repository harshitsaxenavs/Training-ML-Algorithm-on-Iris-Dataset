```python
df = pd.read_excel('iris-train.xlsx')  
```

```python
X = df.drop("class",axis=1) # All columns except the last as features
y = df["class"]   # Last column as label
```

```python
model = RandomForestClassifier()
model.fit(X, y)
```

```python
df_test = pd.read_excel('iris-test.xlsx')  # Replace with your actual CSV file path
```

```python
X_test =  df_test.drop("class",axis=1) 
```

```python
predictions = model.predict(X_test)  # Predict on the same/new data (for demonstration)
print(predictions)  # Display predictions
```

```python
y_test = df_test["class"]
```
